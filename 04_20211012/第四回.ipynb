```python
def rpn(formula: str):
    stack = []
    op = {
        '+': lambda a, b: a+b,
        '-': lambda a, b: a-b,
        '*': lambda a, b: a*b,
        '/': lambda a, b: a/b,
    }
    for token in formula.split():
        if token == "=":
            return stack[-1]
        if token in op:
            y, x = stack.pop(), stack.pop()
            stack.append(op[token](x, y))
        else:
            stack.append(int(token))
    return ' '.join(map(str, stack))
```

```python
print(rpn("3 =") + 5)
```

```python
while True:
    pair = tuple(sorted(map(int, input().split())))
    if pair == (0, 0):
        break
    print(*pair)
```

```python
operators = {
    '+': lambda a, b: a+b,
    '-': lambda a, b: a-b,
    '*': lambda a, b: a*b,
    '/': lambda a, b: a//b,
}
while True:
    _a, _op, _b = input().split()
    if _op not in operators:
        break
    print(operators[_op](*map(int, (_a, _b))))
```
